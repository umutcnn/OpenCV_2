```python
import numpy as np
import matplotlib.pyplot as plt#resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline 
```

```python
from PIL import Image
```

```python
#resmi aç
pic=Image.open('aircraft.jpg')
pic
```

```python
#resmin tipi için
type(pic)
```

```python
#fotograf array donusturme
pic_arr= np.asarray(pic)
```

```python
#resmin tipi için
type(pic_arr)
```

```python
#kaça kaç
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
#kopyasını oluştur
pic_copy = pic_arr.copy()
```

```python
pic_copy.shape
```

```python
#sadece kırmızı olan bölümü yazdir
pic_copy[:,:,0]
```

```python
plt.imshow(pic_copy[:,:,0])
```

```python
#gri tona çevir. kırmızı alanlar genelde beyaza daha yakın olucak
plt.imshow(pic_copy[:,:,0],cmap='gray')
```

```python
#gri tona çevir. yeşil alanlar genelde beyaza daha yakın olucak
plt.imshow(pic_copy[:,:,1],cmap='gray')
```

```python
#gri tona çevir. mavi alanlar genelde beyaza daha yakın olucak
plt.imshow(pic_copy[:,:,2],cmap='gray')
```

```python
#kırmızı ve yeşil değerlere sıfır vererek mavi yoğunluklu bir görsel elde edicez
pic_copy[:,:,0]=0
pic_copy[:,:,1]=0
plt.imshow(pic_copy)
```
