```python
img=cv2.imread('aircraft.jpg')
```

```python
matloplib rgb formatındaydı ancak OpenCV de BGR formatındadır bu nedenle cv2 kütüphanesinde bir düzenleme yaparak onda da 
görselleri rgb formatına getireceğim
```

```python
#rgb formatına dönüştür ve ekrana bas
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
#fotografı manuel boyutlandırma
new_resize = cv2.resize(rgb_img,(1800,500))
plt.imshow(new_resize)
```

```python
#fotografı oranlayarak boyutlandırma
w_ratio=0.5
h_ratio=0.5
new_image = cv2.resize(rgb_img,(0,0),rgb_img,w_ratio,h_ratio)
plt.imshow(new_image)
```

```python
new_image.shape
```

```python
#180 derece çevirme 
new_image=cv2.flip(rgb_img,0)
plt.imshow(new_image)
```

```python
#yansımasını alma çevirme 
new_image=cv2.flip(rgb_img,1)
plt.imshow(new_image)
```

```python
#hem yansımasını alma hem de 180 derece çevirme 
new_image=cv2.flip(rgb_img,-1)
plt.imshow(new_image)
```

```python
#görsel kaydetme
cv2.imwrite('yeni_img.jpg', rgb_img)
```
