```python
import cv2
import numpy as np
import matplotlib.pyplot as plt#resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

```python
#bos siyah bir gorsel olustur
bos_foto=np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(bos_foto)
```

```python
# aşagidaki adresi inceleyerek değerleri bulabiliriz
#https://docs.opencv.org/4.x/d6/d6e/group__imgproc__draw.html
```

```python
cv2.putText(bos_foto,text='merhaba dunya',org=(10,500),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=2, color=(255,255,255),thickness=3)
plt.imshow(bos_foto)
```

```python
#poligon seklinde gorseller cizecegim 
#ilk once cokgenin kenarlarını belirleyecegim
noktalar=np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
#boktaların shape
noktalar.shape
```

```python
#foto adı, noktalar, noktaların kapatılıpğ kapatılmaması,renk, kalınlık
cv2.polylines(bos_foto,[noktalar],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(bos_foto)
```
