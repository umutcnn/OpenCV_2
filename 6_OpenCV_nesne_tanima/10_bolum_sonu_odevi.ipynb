```python
import cv2
import numpy as np
import matplotlib.pyplot as plt #resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

# Araba plakası bulalim

```python
img=cv2.imread('car.JPG')
```

```python
def display(img):
    fig=plt.figure(figsize=(10,8))
    ax=fig.add_subplot(111)
    new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(new_img)
```

```python
display(img)
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_russian_plate_number.xml')
```

```python
def detect_plate(img):
    
    #gelen fotomun bir kopyasını oluşturuyorum
    car_img = img.copy()
    
    plate_recs = face_cascade.detectMultiScale(car_img)
    
    #çizelim
    for (x,y,w,h) in plate_recs:
        cv2.rectangle(car_img,(x,y),(x+w,y+h),(255,0,255),10)
    return car_img
```

```python
result=detect_plate(img)
plt.imshow(result)
```
