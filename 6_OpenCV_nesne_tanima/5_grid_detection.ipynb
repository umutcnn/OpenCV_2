```python
import cv2
import numpy as np
import matplotlib.pyplot as plt #resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

```python
flat_chess = cv2.imread('santranc_tahtasi.jpg')
plt.imshow(flat_chess)
```

```python
#fotografın içersindeki köşeleri belirle
found, corners=cv2.findChessboardCorners(flat_chess,(7,7))
```

```python
#belirlediyse true döndürür
found
```

```python
# eger found değeri true ise 
if found:
    print("opencv corner buldu")
else:
    print("opencv corner buldu")
```

```python
corners
```

```python
corners.shape
```

```python
#gorselin kopyasını olusturuyorum
flat_chess_copy =flat_chess.copy()
```

```python
#bulunan cornerleri foto üzerinde cizelim
cv2.drawChessboardCorners(flat_chess_copy,(7,7),corners,found)
```

```python
plt.imshow(flat_chess_copy)
```

```python
dots = cv2.imread('nokta.jpg')
plt.imshow(dots)
```

```python
#fotografın içersindeki köşeleri belirle
found, corners=cv2.findCirclesGrid(dots,(10,10),cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
dots_copy=dots.copy()
```
