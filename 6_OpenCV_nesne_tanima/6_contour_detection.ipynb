```python
import cv2
import numpy as np
import matplotlib.pyplot as plt #resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

```python
img = cv2.imread('6_ders.jpg',0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
extetnal_contours = np.zeros(img.shape)
```

```python
plt.imshow(extetnal_contours)
```

```python
#bu görselin üzeine belirlediğimiz konturları çizelim
for i in range(len(contours)):
    if hierarchy[0][i][3]==-1:
        cv2.drawContours(extetnal_contours,contours,i,255,-1)#cizim islemlerini yapalim
```

```python
plt.imshow(extetnal_contours,cmap='gray')
```

```python
image_internal=np.zeros(img.shape)

for i in range(len( contours)):
    if hierarchy [0][i][3] != -1: #değerleri -1e eşit olmayanı döndür
        cv2.drawContours(image_internal,contours,i,255,-1 )
    
    
plt.imshow(image_internal,cmap='gray')
```
