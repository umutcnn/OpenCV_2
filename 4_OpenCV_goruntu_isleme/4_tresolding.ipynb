```python
import cv2
import numpy as np
import matplotlib.pyplot as plt #resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

```python
img= cv2.imread('gokkusagi.png')
plt.imshow(img)
```

```python
img= cv2.imread('gokkusagi.png',0)
plt.imshow(img, cmap='gray')
```

```python
img.max()
```

```python
img.min()
```

```python
#piksellerin değerleriyle oynuyoruz 127 den küçük değerleri 0, büyük değerleri ise 255 yap
ret, tresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(tresh1,cmap='gray')
```

```python
#yukardaki islemin aynısı fakat siyahların yerini beyaz beyazların yerini siyah alacak
ret, tresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(tresh2,cmap='gray')
```

```python
img= cv2.imread('bulmaca.jpg')
plt.imshow(img)
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
img_1= cv2.imread('sudoku.jpg',0)
plt.imshow(img_1,cmap='gray')
```

```python
show_pic(img_1)
```

```python
ret,th1=cv2.threshold(img_1,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
#treshold maxsimize etmek için adaptiveTreshold fonksiyonu kullanılır
th2=cv2.adaptiveThreshold(img_1,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
th2=cv2.adaptiveThreshold(img_1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,8)
show_pic(th2)
```
