```python
import cv2
import numpy as np
import matplotlib.pyplot as plt #resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

```python
def load_img():
    blank_img=np.zeros((600,600))
    font=cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300),fontFace=font,fontScale=4, color=(255,255,255), thickness=4)
    return blank_img
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img=load_img()
display_img(img)
```

```python
#aşındırma işlemi için bir kernel oluşturduk
kernel=np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
#aşındırma işlemini yaptık
result=cv2.erode(img,kernel,iterations=1)
```

```python
display_img(result)
```

```python
#fotoğrafa gürültü ekleyeceğiz 
img=load_img()
#numpy kütüphanesinin random kütüphanesini kullanarak 600x600 bir tuval oluşturacağım 
white_noise= np.random.randint(low=0,high=2,size=(600,600))
```

```python
white_noise
```

```python
#bu array ı ekrana bas
display_img(white_noise)
```

```python
img.max()
```

```python
white_noise=white_noise*255
```

```python
white_noise
```

```python
display_img(white_noise)
```

```python
#ilk olusturguğum fotoya gürültü ekledim
noise_img=white_noise+img
```

```python
display_img(noise_img)
```

```python
#gürültülerden kurtulmak için morphologyEx fonksiyonu kullanılır
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
display_img(opening)
```

```python
display_img(img)
```

```python
#harflerin içerisine gurultu yerlestiricem 
black_noise= np.random.randint(low=0,high=2,size=(600,600))
```

```python
black_noise
```

```python
black_noise = black_noise * -255
```

```python
black_noise
```

```python
black_noise_img = img +black_noise
black_noise_img
```

```python
#gürültüyü sadece harfler içerisinde olması için 
black_noise_img[black_noise_img==-255]=0
```

```python
black_noise_img.min()
```

```python
display_img(black_noise_img)
```

```python
#harflerin içerisindeki gürültüden kurtulalım 
closing= cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE,kernel)
```

```python
display_img(closing)
```

```python
#MORPH_GRADIENT fonksiyonu ile harfin çevrelerini belirleyerek içerisinin siyah olmasını sağlayacağım
img=load_img()
display_img(img)
```

```python
grandient= cv2.morphologyEx(img, cv2.MORPH_GRADIENT,kernel)
display_img(grandient)
```
