```python
#duvar fotografındaki renk değerlerini sırasıyla ekrana basalım 
bricks=cv2.imread('duvar.png')
img=bricks
color = ('b','g','r')

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('HISTOGRAM')
```
