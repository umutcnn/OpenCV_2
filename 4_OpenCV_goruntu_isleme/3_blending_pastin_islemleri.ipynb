```python
import cv2
import matplotlib.pyplot as plt #resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

```python
#görselleri al 
img1= cv2.imread('aircraft.jpg')
img2= cv2.imread('warning.jpeg')
```

```python
#bgr to rgb
img1_rgb=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
#ekrana bas
plt.imshow(img1_rgb)
```

```python
plt.imshow(img2_rgb)
```

```python
img1_rgb.shape
```

```python
img2_rgb.shape
```

```python
#iki fotografın boyutunu birbirine esitle
img1_rgb= cv2.resize(img1_rgb,(1200,1200))
img2_rgb= cv2.resize(img2_rgb,(1200,1200))
```

```python
plt.imshow(img1_rgb)
```

```python
plt.imshow(img2_rgb)
```

```python
#boyutlarını esitlediğim iki botoyu blend yap yani üst üste ekle
blended=cv2.addWeighted(src1=img1_rgb, alpha=0.5,src2=img2_rgb,beta=0.5,gamma=0)
plt.imshow(blended)
```

```python
img1= cv2.imread('aircraft.jpg')
img2= cv2.imread('warning.jpeg')
img1_rgb=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2_rgb= cv2.resize(img2_rgb,(600,600))
```

```python
plt.imshow(img1_rgb)
```

```python
plt.imshow(img2_rgb)
```

```python
large_img=img1_rgb
small_img=img2_rgb
```

```python
#görselleri üst üste yapıştır
large_img[0:600,0:600]=small_img
```

```python
plt.imshow(large_img)
```

```python
roi=img1_rgb[0:720,0:1280]
plt.imshow(roi)
```

```python
#gri tonlamalı yap
img2gray=cv2.cvtColor(img2_rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray,cmap='gray')
```

```python
#arka plan siyah yap
mask_inv=cv2.bitwise_not(img2gray)
plt.imshow(mask_inv)
```

```python
mask_inv.shape
```

```python
import numpy as np
```

```python
#gri formattaki görsel ile eşitleyeceğim görselin piksellerini esitlicem
white_backround=np.full((600,600,3),255,dtype=np.uint8)
```

```python
white_backround.shape
```

```python
bk = cv2.bitwise_or(white_backround,white_backround,mask= mask_inv)
```

```python
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg=cv2.bitwise_or(img2_rgb,img2_rgb,mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi=cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```
