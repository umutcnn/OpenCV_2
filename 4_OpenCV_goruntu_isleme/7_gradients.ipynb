```python
import cv2
import numpy as np
import matplotlib.pyplot as plt #resimler üzerinde çizim yapmak için
#çizim yapılan resimleri kaydetmek için
%matplotlib inline
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
# laplacian, sobelx, sobely işlemlerini açıklayan bir görsel ekleyelim 
img=cv2.imread('7_gradients.PNG')
display_img(img)
```

```python
img=cv2.imread('sudoku2.jpg')
display_img(img)
```

```python
#sobel x
sobelx= cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
display_img(sobelx)
```

```python
#sobel y
sobely= cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
display_img(sobely)
```

```python
laplacian=cv2.Laplacian(img,cv2.CV_64F)
display_img(laplacian)
```

```python
#blend yapalım 
blended=cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0)
display_img(blended)
```

```python
#thresh yapalım 
ret,th1=cv2.threshold(img,50,255,cv2.THRESH_BINARY)
display_img(th1)
```

```python
#morphologyEX uygulayalım
kernel=np.ones((4,4),np.uint8)
gradient=cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
display_img(gradient)
```
