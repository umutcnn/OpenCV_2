```python
import cv2

#webcam bağlanma
cap=cv2.VideoCapture(0)

#genişlik
width= int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))

#yukseklik
height= int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#çift / koyarak int değerler aldım ve bu işlemi görüntümün yarısında bir kare olusturmak için yaptım
x=width//2
y=height//2

w=width//4
h=height//4

#bağlandığım kamera goruntusunu ekrana bas
while True:
    #frameleri oku
    ret, frame = cap.read()
    
    #dikdortgen ciz
    cv2.rectangle(frame,(x,y),(x+w,y+h), color=(0,0,255),thickness=4)
    
    #ekrana bas
    cv2.imshow('frame', frame)
    
    #goruntuden çıkmak için
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
   
        
    
#webcam bağlantısı kopar
cap.release()
#ekran kapat
cv2.destroyAllWindows()
```

```python
import cv2

#dikdörtgen çizdirdiğim fonksiyon
def dikdortgen_ciz(event,x,y,flag,param):
    global pt1,pt2,topLeft_clicked,botRight_clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        if topLeft_clicked == True and botRight_clicked == True:
            pt1=(0,0)
            pt2=(0,0)
            topLeft_clicked=False
            botRight_clicked=False
            
        elif topLeft_clicked==False:
            pt1=(x,y)
            topLeft_clicked=True
            
        elif botRight_clicked==False:
            pt2=(x,y)
            botRight_clicked=True
    
    

#global degiskenlerim
pt1 = (0,0)
pt2 = (0,0)
topLeft_clicked = False
botRight_clicked = False


#webcam bağlanma
cap = cv2.VideoCapture(0)

#set mause call back islemi
cv2.namedWindow('test')
cv2.setMouseCallback('test', dikdortgen_ciz)



#bağlandığım kamera goruntusunu ekrana bas
while True:
    #frameleri oku
    ret, frame = cap.read()
    
    #sekilleri cizdirme
    
    #sol tık olduğunda daire ciz
    if topLeft_clicked == True: 
        #daire ozellikleri
        cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),thickness=-1)
        
    if topLeft_clicked and botRight_clicked :
        #dikdortgen ciz
        cv2.rectangle(frame,pt1,pt2,(0,0,255),4)
    
    
    #ekrana bas
    cv2.imshow('test', frame)
    
    #goruntuden çıkmak için
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
   
        
    
#webcam bağlantısı kopar
cap.release()
#ekran kapat
cv2.destroyAllWindows()
```
