```python
import cv2

#dikdörtgen çizdirdiğim fonksiyon
def draw_circle(event,x,y,flag,param):
    global center,clicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        center=(x,y)
        clicked=False
        
        
    if event == cv2.EVENT_LBUTTONUP:
            clicked=True
        
    
    

#global degiskenlerim
center = (0,0)
clicked = False


#webcam bağlanma
cap = cv2.VideoCapture(0)

#set mause call back islemi
cv2.namedWindow('test')
cv2.setMouseCallback('test', draw_circle)



#bağlandığım kamera goruntusunu ekrana bas
while True:
    #frameleri oku
    ret, frame = cap.read()
    
    #sekilleri cizdirme
    
    #sol tık olduğunda daire ciz
    if clicked == True: 
        #daire ozellikleri
        cv2.circle(frame,center=center,radius=50,color=(255,0,0),thickness=5)
        
    
    
    
    #ekrana bas
    cv2.imshow('test', frame)
    
    #goruntuden çıkmak için
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
   
        
    
#webcam bağlantısı kopar
cap.release()
#ekran kapat
cv2.destroyAllWindows()
```
